# Importación de librerías principales y ajustes de visualización

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ajustes de estilo para gráficos más profesionales
plt.style.use('ggplot')
sns.set_context('notebook')
np.random.seed(42)
```

# Carga del dataset

```python
df = pd.read_csv('../data/raw/creditcard.csv')
print(f'Shape del dataset: {df.shape}')
df.head()
```

# Información general y estadísticas descriptivas

```python
df.info()
df.describe().T
```

```python
def resumen_nulos(df):
    """
    Muestra el número y porcentaje de valores nulos por columna en un DataFrame.
    """
    nulos = df.isnull().sum()
    porcentaje = 100 * nulos / len(df)
    resumen = pd.DataFrame({'Nulos': nulos, 'Porcentaje': porcentaje})
    resumen = resumen[resumen['Nulos'] > 0].sort_values('Nulos', ascending=False)
    if resumen.empty:
        print("No hay valores nulos en el DataFrame.")
    else:
        display(resumen)
        
resumen_nulos(df)
```

# Distribución de la variable 'Class'

```python
fraud_count = df['Class'].value_counts()
print('Conteo de clases:\n', fraud_count)

plt.figure(figsize=(6,4))
sns.countplot(x='Class', data=df, palette='coolwarm')
plt.title('Distribución de la variable Class (0: No Fraude, 1: Fraude)')
plt.show()

fraud_ratio = fraud_count[1] / fraud_count.sum()
print(f'Porcentaje de transacciones fraudulentas: {fraud_ratio:.4%}')
```

El dataset está altamente desbalanceado: la clase positiva (fraude) representa solo el 0.172% de las transacciones totales.

Este desbalance tan extremo justifica que no podemos usar métricas como la exactitud (accuracy) para evaluar el modelo. Debemos usar métricas más robustas como ROC AUC, PR AUC y F1-score.

# Distribución de 'Amount' y 'Time'

```python
plt.figure(figsize=(8,4))
sns.histplot(df['Amount'], bins=50, kde=True, color='dodgerblue')
plt.title('Distribución de Amount')
plt.xlabel('Monto')
plt.ylabel('Frecuencia')
plt.show()

plt.figure(figsize=(8,4))
sns.histplot(df['Time'], bins=50, kde=True, color='green')
plt.title('Distribución de Time (segundos desde la primera transacción)')
plt.xlabel('Tiempo (segundos)')
plt.ylabel('Frecuencia')
plt.show()
```

**Distribución de la variable Amount:**

La mayoría de las transacciones tienen montos bajos. Existe una cola larga hacia montos muy elevados.

Dado que el monto puede afectar la decisión de fraude, probablemente sea útil normalizar esta variable o aplicarle log-transformaciones.

**Distribución de la variable Time:**

Hay dos picos claros a lo largo del tiempo, lo que podría reflejar patrones de transacción por franjas horarias.

No parece haber un patrón claro entre el tiempo y la clase, pero puede aportar información en combinación con otras variables.

# Correlación de las features con 'Class'

```python
corr_matrix = df.corr()

plt.figure(figsize=(12,8))
corr_with_target = corr_matrix['Class'].drop('Class')
corr_with_target.sort_values().plot(kind='barh', color='steelblue')
plt.title('Correlación de las features con la clase (fraude)')
plt.show()
```

La magnitud de las correlaciones es moderada, lo que sugiere que ningún predictor por sí solo es suficientemente fuerte, pero en conjunto podrían aportar valor.

# Matriz de Correlación de las features

```python
plt.figure(figsize=(20,20))
sns.heatmap(corr_matrix, cmap='coolwarm', center=0, annot=True, fmt=".2f", linewidths=0.5, annot_kws={"size":8})
plt.title('Matriz de correlación de todas las variables')
plt.show()
```

La mayoría de las variables están poco correlacionadas entre sí, excepto algunas como V1-V2 o V13-V15.

Esto es bueno para los modelos, ya que la multicolinealidad no parece un gran problema.

# Boxplots de features principales (sin outliers)

```python
pca_features = [f'V{i}' for i in range(1, 29)]

plt.figure(figsize=(15,8))
df[pca_features].boxplot(showfliers=False)
plt.title('Distribución de las variables PCA (sin outliers)')
plt.xticks(rotation=90)
plt.show()
```

Las variables transformadas por PCA están centradas en 0 y presentan distribuciones similares.

No hay valores extremos claros después de la transformación, pero las colas en algunos casos indican cierta dispersión.

# Estadísticas de Amount por clase

```python
print(df.groupby('Class')['Amount'].describe())

# Comparación de distribuciones de Amount por clase
plt.figure(figsize=(8,4))
sns.kdeplot(df.loc[df['Class']==0, 'Amount'], label='No Fraude', shade=True)
sns.kdeplot(df.loc[df['Class']==1, 'Amount'], label='Fraude', shade=True, color='red')
plt.title('Distribución de Amount por clase')
plt.legend()
plt.show()
```

Las transacciones fraudulentas tienden a ser de montos menores a las no fraudulentas (aunque esto debe confirmarse con pruebas estadísticas).

Esto podría indicar que los fraudes se concentran en montos bajos para pasar desapercibidos.

# Transformacion de variables

```python
# convertir time en horas del día
df["Hour"] = (df["Time"] / 3600) % 24
```

```python
# Distribución de Hour para fraudes y no fraudes
plt.figure(figsize=(10,5))
sns.kdeplot(df[df['Class']==0]['Hour'], label='No Fraude', shade=True)
sns.kdeplot(df[df['Class']==1]['Hour'], label='Fraude', shade=True, color='red')
plt.title('Distribución de Hour por clase')
plt.xlabel('Hour')
plt.legend()
plt.show()
```

La distribución de Hour muestra diferencias visibles entre transacciones de fraude y no fraude.

Fraude (línea roja): Tiende a concentrarse en las primeras horas del día (picos entre las 2-8 horas).

No fraude (línea azul): Presenta una distribución más uniforme durante el día (picos alrededor de las 12-20 horas).

Esto sugiere que hay patrones horarios relevantes para identificar fraudes.
